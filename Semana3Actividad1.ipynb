Parte 1: Fundamentos de bases de datos
Las bases de datos son una herramienta fundamental en todas las industrias, apoyan con la persistencia de datos por una parte, pero aún más importante es que relacionan la información, dandole así cierta lógica y un sinfin de usos. Existen las relacionales (SQL) y no relacionales (no llevan un mismo orden como el caso de MongoDB). Normalmente las bases se componen de tablas que representan un componente con determinados atributos o columnas y cada uno de los registros contenidos en estas tablas es una fila. Es posible utilizar operacions DML para modificar o consultar el contenido de las tablas, o incluso para realizar operaciones que involucren varias tablas. Una manera común de representar estas interacciones es a través de un diagrama ERD (Entidad Relación).
En cuanto a la limpieza de datos, es sumamente importante realizar algún tipo de preprocesamiento si los datos se encuentran en crudo Si tenemos un conjunto de datos con basura, seguramente las predicciones de los módelos contendrán basura también. Un ejemplo de lipieza que se debe hacer es obtener los datos faltantes o modelarlos si es que se da el tipo de MNAR, esto se puede llevar a cabo a través de eliminación por fila, por pares, se sustituyen los espacios en blanco por valor medio del conjunto o incluso se puede realizar una sustituciób de los valores por algunos valores que aparentemente serían correctos. En cuanto al ruido que también se debe limpiar existen herramientas como el box plot o agrupación donde al identificar limites, outliers o comportamientos intercuartiles se pueden ir haciendo ajustes para suavizar los resultados finales del modelo.

###Las columnas representan lo siguiente
This research employed a binary variable, default payment (Yes = 1, No = 0), as the response variable. This study reviewed the literature and used the following 23 variables as explanatory variables: 
X1: Amount of the given credit (NT dollar): it includes both the individual consumer credit and his/her family (supplementary) credit. 
X2: Gender (1 = male; 2 = female). 
X3: Education (1 = graduate school; 2 = university; 3 = high school; 4 = others). 
X4: Marital status (1 = married; 2 = single; 3 = others). 
X5: Age (year). 
X6 - X11: History of past payment. We tracked the past monthly payment records (from April to September, 2005) as follows: X6 = the repayment status in September, 2005; X7 = the repayment status in August, 2005; . . .;X11 = the repayment status in April, 2005. The measurement scale for the repayment status is: -1 = pay duly; 1 = payment delay for one month; 2 = payment delay for two months; . . .; 8 = payment delay for eight months; 9 = payment delay for nine months and above. 
X12-X17: Amount of bill statement (NT dollar). X12 = amount of bill statement in September, 2005; X13 = amount of bill statement in August, 2005; . . .; X17 = amount of bill statement in April, 2005. 
X18-X23: Amount of previous payment (NT dollar). X18 = amount paid in September, 2005; X19 = amount paid in August, 2005; . . .;X23 = amount paid in April, 2005.

```python
#Pruebas de limpieza por imputacion
df5 = df.copy()
df5.describe()
```

```python
df5.isnull().any()
```

```python
#Se inicia la limpieza con una nueva copia
df5 = df.copy()
#Imputation para X2:
#Usareos moda ya que solo hay 2 valores
modeX2 = df5.X2.mode()[0]
df5['X2'].fillna(value=modeX2, inplace=True)
#Imputation para X3:
#Usareos moda ya que solo hay ciertos valores
modeX3 = df5.X3.mode()[0]
df5['X3'].fillna(value=modeX3, inplace=True)
#Imputation para X4:
#Usaremos moda ya que solo hay ciertos valores
modeX4 = df5.X4.mode()[0]
df5['X4'].fillna(value=modeX4, inplace=True)
#Imputation para X3:
#Usaremos media ya que la edad puede arrojar valores muy variados
meanX5 = np.round(df5.X5.mean()) #Rounding the value with Numpy Round
df5['X5'].fillna(value=meanX5, inplace=True)
#Hagamos el drop de filas con valores vacios entre las columnas X6-X23.
#Revisemos cuantas filas faltan
X6to23 = list(df5.iloc[:,6:24].columns) #Obteniendo colunas X6 a X23
print('Dropping from columns',','.join(X6to23)) 
df5.dropna(subset=X6to23, inplace=True) #Eliminando valores vacios en el intervalo

oRows = df.shape[0]
nRows = df5.shape[0]
per = 100 - ((nRows * 100) / oRows)
print("Original Rows:",oRows)
print("After Imputed and Dropped Rows:",nRows)
print("Percentage of rows lost: {}%".format(per))

print("Are there Null rows after cleaning:",df5.isnull().values.any())
```
